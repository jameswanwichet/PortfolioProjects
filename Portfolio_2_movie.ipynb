```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) 

df = pd.read_csv('/Users/jameswanwichet/VScode/Projects/Dataset/movies.csv')
```

```python
# Look at the data

df.head()
```

```python
# Any missing data

for col in df.columns: 
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns
df.dtypes
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
pd.set_option('display.max_rows', 20)
```

```python
# Drop any duplicates
df.drop_duplicates()
```

```python
#df_cleaned = df.dropna()
df_cleaned = df.fillna(0)
print(df_cleaned.isna().sum())
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df_cleaned['budget'], y=df_cleaned['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df_cleaned, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
df_cleaned.corr(method='pearson') #pearson, kendall, spearman
```

```python
# High correlation between bedget and gross

correlation_matrix = df_cleaned.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation  Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Looks at Company

df_numerized = df_cleaned
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
print(df_numerized)
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation  Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
print(high_corr)
```
